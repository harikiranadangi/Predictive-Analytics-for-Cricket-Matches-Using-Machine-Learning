### **Creating Date Column**

```python
# Clean up the dates column
df['dates'] = df['dates'].str.strip("[]").str.replace("'", "").str.strip()

# Extract the latest date in cases where there are multiple dates
def extract_latest_date(date_str):
    dates = date_str.split(', ')
    if len(dates) > 1:
        return dates[-1]
    else:
        return dates[0]

df['dates'] = df['dates'].apply(extract_latest_date)

# Convert the dates to datetime format
df['dates'] = pd.to_datetime(df['dates'], errors='coerce')

# Sort DataFrame by dates
df = df.sort_values(by="dates", ascending=False)

df
```

```python
df['city'] = df['city'].fillna('')

# Concatenate 'ground' and 'city' columns
df.loc[:, 'ground'] = df['ground'] + "," + df['city']

# Remove any leading or trailing commas and spaces
df['ground'] = df['ground'].str.strip(', ')
```

```python
# Drop the 'city' column
df = df.drop(columns="city")
```

```python
def clean_ground_name(ground):
    if isinstance(ground, str):
        parts = ground.rsplit(',', 1)  # Split by the last comma
        if len(parts) == 2 and parts[0].endswith(parts[1]):
            return parts[0]
    return ground

def add_space_after_last_comma(value):
    if isinstance(value, str):
        last_comma_index = value.rfind(',')
        if last_comma_index != -1:
            value = value[:last_comma_index + 1] + ' ' + value[last_comma_index + 1:].strip()
    return value


# Apply the function to the 'ground' column
df['ground'] = df['ground'].apply(clean_ground_name)
df['ground'] = df['ground'].apply(add_space_after_last_comma)

# Display the modified DataFrame
df
```
